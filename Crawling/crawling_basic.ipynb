```python
import requests
from bs4 import BeautifulSoup
```

# get

```python
url = 'https://comic.naver.com/webtoon/weekdayList.nhn?week=mon'
res = requests.get(url=url)
print(res)
    # Response [200] 이면 성공
    
soup = BeautifulSoup(res.text,'html.parser')
print(soup)
```

## 크롬과 bs4 차이
크롬 : content > div.list_area.daily_img > ul > li:nth-child(1) > dl > dt > a

bs4 : content > div.list_area.daily_img > ul > li:nth**-of-type**(1) > dl > dt > a

```python
title = soup.select('div.list_area.daily_img > ul > li:nth-of-type(1) > dl > dt > a')
print(title)
print(title[0].text)
```

```python
titles = soup.select('div.list_area.daily_img > ul > li > dl > dt > a')
print([x.text for x in titles])
```

```python
print([x.get('title') for x in titles])
```

## 실시간 검색어

웹사이트를 불러올 때 순서대로 나옴

-> 즉 수많은 get 명령을 통해 데이터를 불러옴

-> 그래서 빈 list가 반환되는 경우도 있음

```python
# 실시간 검색어
url = 'https://apis.naver.com/mobile_main/srchrank/srchrank?frm=main&ag=20s&gr=4&ma=-1&si=2&en=0&sp=1'
res = requests.get(url=url)
soup = BeautifulSoup(res.text, 'html.parser')
title = soup.select('NM_RTK_VIEW_list_wrap > ul:nth-child(2) > li:nth-child(2) > a > span')
print(title)
    # 수많은 get으로 인해 빈 list 반환
```

```python
url = 'https://apis.naver.com/mobile_main/srchrank/srchrank?frm=main&ag=20s&gr=4&ma=-1&si=2&en=0&sp=1'
res = requests.get(url=url)
res = res.json()
title = res['data']
print([x['keyword'] for x in title])
    # Response [200] 이면 성공
```

## crawling을 제한한 경우

Request header를 추가 하면 됨

userid 등 여러가지 하나씩 추가하다 보면 됨

```python
url = 'https://www.reddit.com/r/datascience/'
res = requests.get(url=url)
print(res)
```

```python
url = 'https://www.reddit.com/r/datascience/'
headers = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36'
}
res = requests.get(url=url, headers=headers)
print(res)
```

### headers와 params
headers : request-get-header

params : url 주소 ? 뒤에 해당하는 parameters

```python
url = 'https://search.naver.com/search.naver'
params = {
    'query':'데이터사이언스랩'
}
res = requests.get(url=url, params=params)
print(res)
print(res.text)
```

# post
params와 비슷한 것을 data를 통해 전달

```python
url = 'http://kiss.kstudy.com/search/sch-result.asp'
data = {
    'startCount': 0,
    'pageScale': 10,
    'json': '{"prefixQuery":{"inst":null,"publ":null,"exYear":null,"issn":null,"mbcdls":[],"publPldv":null,"ciInfo":null},"collectionQuery":{"queries":[{"field":"전체","value":"박상언","prefix":""}],"reQueries":[]},"dateRange":null,"sort":null,"isForwardMatch":false}'
}
res = requests.post(url=url, data=data).text
soup = BeautifulSoup(res, 'html.parser')
temp = soup.select('div.list-title.is1 > span.count')
print(temp)
print(temp[0].text)
```

## proxies

## multiprocessing

```python
from multiprocessing import Pool
```

```python
# 추후 오류 확인해 보기

def crawler(url):
    html = requests.get(url).text
    soup = BeautifulSoup(html, 'html.parser')
    print(soup.select('h1')[0].text) # 첫 h1 태그

if __name__ == '_main_':
    pool = Pool(processes=10)
    output = pool.map(crawler, 'http://www.naver.com')
    print(output)
```
