# 과제
좋아하는 연예인의 이름을 네이버에 검색하여 뉴스헤드라인만 100개 긁어오기.

```python
import math
import time
from tqdm.notebook import tqdm
import requests
from bs4 import BeautifulSoup
```

```python
title=[]
```

```python
n = 100
query = '아린'
url = 'https://search.naver.com/search.naver'

title=[]

for cnt in tqdm(range(math.ceil(n/10))):
    params = {
        'where': 'news',
        'query': query,
        'start': cnt*10-9
    }
    
    res = requests.get(url=url, params=params).text
    soup = BeautifulSoup(res, 'html.parser')
    temp = soup.select('a._sp_each_title')
    for i in range(10):
        title.append(temp[i].text)
        
    time.sleep(1)
```
